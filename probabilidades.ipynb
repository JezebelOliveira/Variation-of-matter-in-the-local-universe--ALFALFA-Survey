```python
import numpy as np
import matplotlib.pyplot as plt

# Definindo os parmetros da distribuio normal
mu = np.array([0.78, 0.80, 0.83,0.811,0.819])  # mdias
sigma = np.array([0.04, 0.05, 0.05,0.006,0.015])  # desvios padro
num_samples = 10000  # nmero de amostras a serem geradas para cada distribuio
h = np.array([0.6727,0.698,0.7304,"Planck","ACT"])

# Gerando amostras aleatrias com distribuio normal para cada conjunto de mdia e desvio padro
samples = []
for i in range(len(mu)):
    samples.append(np.random.normal(mu[i], sigma[i], num_samples))



# Plotando a funo de densidade de probabilidade terica para cada distribuio
plt.figure(figsize=(10, 6))
for i in range(len(mu)):
    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 100000)
    p = np.exp(-0.5 * (x - mu[i])**2 / sigma[i]**2) / (sigma[i] * np.sqrt(2 * np.pi))
    if i == 3:
        plt.plot(x, p, linewidth=2, label=h[i])  # Plotando para "Planck"
        
    elif i == 4:
        plt.plot(x, p, linewidth=2, label=h[i])  # Plotando para "ACT"
        plt.plot(x, p, linewidth=2, label=f'h={h[i]}')

plt.xlabel(f'$\sigma_8$',fontsize=16)
plt.ylabel('Densidade de Probabilidade',fontsize=16)
plt.tick_params(labelsize=14)

plt.legend(loc=1)
plt.xlim(0.65,0.95)
plt.savefig("probabilidade.pdf")
```

```python
# Definindo os valores de sigma_8 do Planck e do ACT
sigma8_planck = 0.811
sigma8_act = 0.819

# Calculando em quantos sigmas as medies simuladas esto dos valores do Planck e do ACT
sigma_diff_planck = np.abs(mu - sigma8_planck) / sigma
sigma_diff_act = np.abs(mu - sigma8_act) / sigma
```
