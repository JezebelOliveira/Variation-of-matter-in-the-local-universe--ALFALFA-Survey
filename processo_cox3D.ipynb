```python
#carregando as principais bibliotecas

import numpy as np
import matplotlib.pyplot as plt
from astropy.coordinates import cartesian_to_spherical
```

```python
#definindo parâmetros

l = 25      
L = 200
V = L**3
Lv = 0.008

Ns = np.int64((Lv * V) / l)

sig = 0.1


#Definindo o tamanho do cubo 

N = 1000
for j in range(N):

    XI = []
    YI = []
    ZI = []
    for i in range(Ns):
        
        x0 = np.random.uniform(-L/2., +L/2., 1)
        y0 = np.random.uniform(-L/2., +L/2., 1)
        z0 = np.random.uniform(-L/2., +L/2., 1)

        #vetor unitário

        R     = 1.
        alfa  = np.random.uniform(0., 2.*np.pi, 1)
        delta = np.random.uniform((-np.pi/2), (np.pi/2), 1)       
        
        v1 = R * np.cos(alfa) * np.sin(delta)
        v2 = R * np.sin(alfa) * np.sin(delta)
        v3 = R * np.cos(delta) 

                
        ls = 1.8
        Np = ls * l
        
        t = np.random.uniform(1, l, np.int64(Np))
        #t = np.arange(1, l, 1)
        
        X = x0 + t*v1
        Y = y0 + t*v2
        Z = z0 + t*v3
    
        XI.append(x0)
        XI.append(X)
        
        YI.append(y0)
        YI.append(Y)
        
        ZI.append(z0)
        ZI.append(Z)
    
    X = np.array(XI, dtype='object')
    Y = np.array(YI, dtype='object')
    Z = np.array(ZI, dtype='object')
    
    X1 = np.concatenate(X)
    Y1 = np.concatenate(Y)
    Z1 = np.concatenate(Z)
    
    X2 = np.random.normal(X1, sig)    
    Y2 = np.random.normal(Y1, sig)
    Z2 = np.random.normal(Z1, sig)
    
    Xi = []
    Yi = []
    Zi = []
    for i in range(len(X2)):
        
        if -L/2.<=X2[i]<=+L/2. and -L/2.<=Y2[i]<=+L/2. and -L/2.<=Z2[i]<=L/2.:
            Xi.append(X2[i])
            Yi.append(Y2[i])
            Zi.append(Z2[i])
    
    Xi = np.array(X2)
    Yi = np.array(Y2)
    Zi = np.array(Z2)    
       
    r, lat, lon = cartesian_to_spherical(Xi, Yi, Zi)
    
    r   = np.array(r)
    lat = np.array(lat)
    lon = np.array(lon)
       
    # print(len(Xi))

  #salvando os resultados   
    N = r, np.rad2deg(lon), np.rad2deg(lat), Xi, Yi, Zi

    np.savetxt('mocks/mock_'+str(j)+'.dat', np.transpose(N), 
                fmt='%.6f', delimiter='\t')
```
